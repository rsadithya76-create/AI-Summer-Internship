```python
import streamlit as st
from transformers import pipeline
import time

generator = pipeline(
    "text-generation",
    model="distilgpt2"
)

st.title("📝 AI Poem Generator")

st.header("Generate Beautiful Poems using AI")

st.markdown("""
Enter a topic and customize the poem settings below.
""")

st.sidebar.markdown("# Settings")

st.sidebar.info("Adjust the poem generation parameters.")

st.sidebar.write("Developed using Streamlit and Hugging Face")

topic = st.text_area(
    "Enter a Topic",
    height=200
)

max_len = st.slider(
    "Poem Length",
    min_value=50,
    max_value=200,
    value=100,
    step=10
)

num_poems = st.selectbox(
    "Number of Poems",
    options=[1, 2, 3],
    index=0
)

if st.button("Generate Poem"):

    prompt = f"Write a beautiful poem about {topic}"

    with st.spinner("Generating poem..."):

        time.sleep(2)

        result = generator(
            prompt,
            max_length=max_len,
            num_return_sequences=num_poems
        )

    st.markdown("---")

    st.subheader("Generated Poem")

    for i, poem in enumerate(result, start=1):
        st.write(f"### Poem {i}")
        st.write(poem['generated_text'])
```
