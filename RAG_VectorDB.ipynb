```python
from sentence_transformers import SentenceTransformer
import faiss
import groq
```

```python
client=groq.Client(api_key="")
```

```python
docs=[
    "machine leraning is a subfield of artifical intelligence",
    "artifical intelligence is  taking pllace in everywhere",
    "since i was a  AI student i know how to surive in this ai  era",
    "data is a powerfull source for this date than petrol"
]
```

```python
model=SentenceTransformer("sentence-transformers/multi-qa-MiniLM-L6-cos-v1")
doc_embeddings=model.encode(docs).astype("float32")
```

```python
doc_embeddings.shape
```

```python
dimension=doc_embeddings.shape[1]
index=faiss.IndexFlatL2(dimension)   #similar words are grouped using eculidian distance
index.add(doc_embeddings)            # embedded value is also stored along with it
```

```python
query="what is ai"
query_embedding=model.encode([query]).astype("float32")

top_k=2
distance ,indices =index.search(query_embedding,top_k)
ret_ch=[docs[i]for i in indices[0]]
print("retrieved chunks:")
for ch in ret_ch:
    print("-",ch)
```

```python
context="\n".join(ret_ch)
prompt_content=f"""
Answer the question based on the context below
context:
{context}
question:
{query}
answer:
"""

response=client.chat.completions.create(
    model="openai/gpt-oss-20b",
    messages=[
        {"role":"user", "content": prompt_content}
    ],
    temperature=0.2
)
print("Final Answer:\n")
print(response.choices[0].message.content)
```
