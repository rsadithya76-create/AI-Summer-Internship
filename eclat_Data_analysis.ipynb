```python
transactions=[]
for i in range(0,len(df)):
  transactions.append(str(df.values[i,j]) for j in range(0,20) if df.values[i,j]!='nan')
transactions
```

```python
from itertools import combinations

item_tidset = {}

for tid, transaction in enumerate(transactions):
    for item in transaction:
        if item not in item_tidset:
            item_tidset[item] = set()
        item_tidset[item].add(tid)
```

```python
min_support_count = 20
```

```python
frequent_itemsets = []

items = list(item_tidset.keys())

for item1, item2 in combinations(items, 2):

    common_tidset = item_tidset[item1] & item_tidset[item2]

    support_count = len(common_tidset)

    if support_count >= min_support_count:

        frequent_itemsets.append([
            item1,
            item2,
            support_count,
            support_count / len(transactions)
        ])
```

```python
df_eclat = pd.DataFrame(
    frequent_itemsets,
    columns=[
        'Item 1',
        'Item 2',
        'Support Count',
        'Support'
    ]
)

df_eclat
```

```python
df_eclat = df_eclat.sort_values(
    by='Support',
    ascending=False
)

print(df_eclat.head(20))
```
