```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv('/content/Market_Basket_Optimisation.csv' ,header=None)
```

```python
df
```

```python
df.isnull().sum()
```

```python
transactions=[]
for i in range (0,len(df)):

 transactions.append([str(df.values[i,j])for j in range (0,20)if str(df.values[i,j])!='nan'])
```

```python
!pip install apyori
```

```python
from apyori import apriori
```

```python
rules=apriori(transactions,min_support=0.005,min_confidence=0.2,min_lift=3,min_length=3,max_length=3)
result=list(rules)
result
```

```python
def inspect(results):
    lhs         = [tuple(result[2][0][0]) for result in results]
    rhs         = [tuple(result[2][0][1]) for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences, lifts))

resultsinDataFrame = pd.DataFrame(inspect(result), columns = ['Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift'])
```
