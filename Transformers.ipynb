```python
from transformers import pipeline
classifier=pipeline("sentiment-analysis")
classifier("i ")
```

```python
classifier=pipeline("zero-shot-classification")
classifier("i am going to bus ",candidate_labels=["business","education"])
```

```python
gen=pipeline("text-generation")
gen("today i went to restroom, and water is not coming so")
```

```python
gen=pipeline("text-generation",model="sapientinc/HRM-Text-1B")
gen("i woke up late and hurry to the work",max_new_tokens=50,num_return_sequences=2)
```

```python

unmasker = pipeline('fill-mask', model='bert-base-cased')
sentence = "The [MASK] was stuck in the tree."
results = unmasker(sentence)
```

```python
from transformers import pipeline

ner = pipeline("ner",model='bert-base-cased')

result = ner("Elon Musk is the CEO of Tesla")

print(result)
```

```python
from transformers import pipeline
translator = pipeline("translation", model='Helsinki-NLP/opus-mt-en-fr')

result = translator("I love artificial intelligence")

print(result)
```
